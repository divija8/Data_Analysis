```python
df_combined['start_time'] = pd.to_datetime(df_combined['start_time'])

df_combined['end_time']  = pd.to_datetime(df_combined['end_time'])
```

```python
df_combined['start_station_id']= df_combined['start_station_id'].astype('object')
df_combined['end_station_id']= df_combined['end_station_id'].astype('object')

df_combined['user_type'] = df_combined['user_type'].astype('category')
df_combined['member_gender'] = df_combined['member_gender'].astype('category')
```

```python
numeric_data = df_combined.select_dtypes(include=['number'])
non_numeric_data = df_combined.select_dtypes(exclude=['number'])

correlation_matrix = numeric_data.corr()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


correlation_matrix = numeric_data.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix Heatmap')
plt.savefig('correlation_matrix_heatmap.png')  
plt.show()
```

```python
df_combined['start_date'] = df_combined['start_time'].dt.date
df_combined['end_date'] = df_combined['end_time'].dt.date
```

```python
df_combined.head()
```

```python
subscriber_mask = df_combined['user_type'] == 'Subscriber'
customer_mask = df_combined['user_type'] == 'Customer'

subscriber_counts = df_combined.loc[subscriber_mask].groupby(['start_date', 'end_date']).size().reset_index(name='subscriber_count')
customer_counts = df_combined.loc[customer_mask].groupby(['start_date', 'end_date']).size().reset_index(name='customer_count')

merged_counts = pd.merge(subscriber_counts, customer_counts, on=['start_date', 'end_date'], how='outer').fillna(0)

plt.figure(figsize=(12, 6))
sns.lineplot(data=merged_counts, x='start_date', y='subscriber_count', label='Subscribers')
sns.lineplot(data=merged_counts, x='start_date', y='customer_count', label='Customers')
plt.title('Count of Subscribers and Customers over Time')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
df_combined['start_time'] = pd.to_datetime(df_combined['start_time'])

df_combined['day_week']=df_combined['start_time'].dt.day_name()
df_combined['month_val']=df_combined['start_time'].dt.month_name()
df_combined['dur_mins']=df_combined['duration_sec']/60
df_combined['hour_of_the_day'] = df_combined.start_time.dt.strftime('%H')
df_combined['month_no']=df_combined.start_time.dt.month.astype(int)

#Calculating the age when the drive was taken :
df_combined['age_val']=2018-df_combined['member_birth_year']
```

```python
plt.figure(figsize=(10,6));
bins=np.arange(0,df_combined['dur_mins'].max()+1,1)
xticks=[0,5,10,15,20,25,30,35,40,45,50,55,60]
labels=['{}'.format(v) for v in xticks]
plt.xlim(0,70)
plt.hist(data=df_combined,x='dur_mins',bins=bins);
plt.xticks(xticks,labels);
plt.xlabel('Trip Duration in minutes');
plt.ylabel('count');
plt.suptitle(' FordGo Bike System - Trip Duration in Mins',fontsize=15,fontweight='bold');
```

```python
busiest_hours = df_combined.groupby('hour_of_the_day').size()
busiest_days = df_combined.groupby('day_week').size()

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.barplot(x=busiest_hours.index, y=busiest_hours.values)
plt.xlabel('Hour of Day')
plt.ylabel('Number of Trips')
plt.title('Busiest Hours for Bike Usage')


```

```python
plt.subplot(1, 2, 2)
sns.barplot(x=busiest_days.index, y=busiest_days.values)
plt.xlabel('Day of Week')
plt.ylabel('Number of Trips')
plt.title('Busiest Days for Bike Usage')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
duration_comparison = df_combined.groupby('user_type')['dur_mins'].mean()

plt.figure(figsize=(8, 6))
sns.barplot(x=duration_comparison.index, y=duration_comparison.values)
plt.xlabel('User Type')
plt.ylabel('Average Trip Duration (minutes)')
plt.title('Trip Duration Comparison between Customers and Subscribers')
plt.show()
```
